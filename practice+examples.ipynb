# seaborn

Draw a categorical plot onto a FacetGrid.

The default plot that is shown is a point plot, but other seaborn categorical plots can be chosen with the kind parameter, including box plots, violin plots, bar plots, or strip plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
exercise = sns.load_dataset("exercise")
%matplotlib inline
```

```python
g=sns.factorplot(x="time",y="pulse",hue="kind",data=exercise)
```

```python
g = sns.factorplot(x="time", y="pulse", hue="kind",
                  data=exercise, kind="violin")
```

"col" is separating graphs according to one feature. "hue" encodes another feature.

```python
g = sns.factorplot(x="time", y="pulse", hue="kind",
                  col="diet", data=exercise)
```

```python
g = sns.factorplot(x="time", y="pulse", hue="kind",
                  col="diet", data=exercise, palette='Set2')
```

So palette is the color set.

```python
g = sns.factorplot(x="time", y="pulse", hue="kind",
                  col="diet", size=10, data=exercise)
```

"size" is the size.

```python
 g = sns.factorplot(x="time", y="pulse", hue="kind",
                   col="diet", data=exercise,
                   size=5, aspect=.8)
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
g = sns.factorplot("alive", col="deck", col_wrap=4,
                  data=titanic[titanic.deck.notnull()],
                   kind="count", size=2.5, aspect=.8)
```

There are four graphs in one row.

```python
g = sns.factorplot(x="age", y="embark_town",
                  hue="sex", row="class",
                   data=titanic[titanic.embark_town.notnull()],
                   orient="h", size=2, aspect=3.5, palette="Set3",
                   kind="violin", dodge=True, cut=0, bw=.2)
```

```python
g = sns.factorplot(x="age", y="embark_town",
                  hue="sex", row="class",
                   data=titanic[titanic.embark_town.notnull()],palette="Set3",
                   kind="violin", dodge=True, cut=0, bw=.2)
```

```python
g = sns.factorplot(x="who", y="survived", col="class",
                  data=titanic, saturation=.5,
                 kind="bar", ci=None, aspect=.6)
```

kind is just the kind of the graph

```python
g.set_axis_labels("", "Survival Rate").set_xticklabels(["Men", "Women", "Children"]).set_titles("{col_name} {col_var}").set(ylim=(0, 1)).despine(left=True)  
```
